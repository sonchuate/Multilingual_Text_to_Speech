```python
# import os
# !pip install -q pydub
# from pydub import AudioSegment
# from pydub.silence import split_on_silence

# def split_audio_on_silence(file_path, output_dir, min_silence_len=350, silence_thresh=-40, keep_silence=500, sample_rate=22050):
#     # Đọc file âm thanh
#     audio = AudioSegment.from_file(file_path, format="mp3")

#     # Tạo thư mục output nếu chưa tồn tại
#     if not os.path.exists(output_dir):
#         os.makedirs(output_dir)

#     # Chia âm thanh dựa trên khoảng im lặng
#     segments = split_on_silence(audio,
#                                 min_silence_len=min_silence_len,
#                                 silence_thresh=silence_thresh,
#                                 keep_silence=keep_silence)

#     # Xuất từng đoạn âm thanh thành file .wav với tần số lấy mẫu 22kHz
#     for i, segment in enumerate(segments):
#         segment = segment.set_frame_rate(sample_rate)
#         output_path = f"{output_dir}/segment_{i}.wav"
#         segment.export(output_path, format="wav")
#         print(f"Exported {output_path}")

import os
!pip install -q pydub
from pydub import AudioSegment
from pydub.silence import split_on_silence
# tạo dữ liệu mono
def split_audio_on_silence(file_path, output_dir, min_silence_len=350, silence_thresh=-40, keep_silence=500, sample_rate=22050):
    """
    Chia file âm thanh thành các đoạn nhỏ dựa trên khoảng lặng và lưu các đoạn này dưới dạng file WAV.

    Parameters:
    - file_path: Đường dẫn đến file âm thanh đầu vào (MP3).
    - output_dir: Thư mục lưu các file âm thanh đầu ra.
    - min_silence_len: Độ dài tối thiểu của khoảng lặng để coi đó là điểm chia (ms).
    - silence_thresh: Ngưỡng âm lượng để coi đó là khoảng lặng (dB).
    - keep_silence: Giữ một phần của khoảng lặng ở đầu và cuối của mỗi đoạn âm thanh (ms).
    - sample_rate: Tần số lấy mẫu cho các file WAV đầu ra (Hz).
    """

    # Đọc file âm thanh
    audio = AudioSegment.from_file(file_path, format="mp3")

    # Tạo thư mục output nếu chưa tồn tại
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Chia âm thanh dựa trên khoảng im lặng
    segments = split_on_silence(audio,
                                min_silence_len=min_silence_len,
                                silence_thresh=silence_thresh,
                                keep_silence=keep_silence)

    # Xuất từng đoạn âm thanh thành file .wav với tần số lấy mẫu 22kHz, kênh đơn, 16-bit PCM
    for i, segment in enumerate(segments):
        segment = segment.set_frame_rate(sample_rate)
        segment = segment.set_channels(1)
        segment = segment.set_sample_width(2)
        output_path = f"{output_dir}/segment_{i}.wav"
        segment.export(output_path, format="wav")
        # print(f"Exported {output_path}")

# # Sử dụng hàm
# file_path = "/content/drive/MyDrive/y2mate.com - Bình thản chấp nhận một bản thân không hoàn hảo  Sách Sức mạnh của cảm xúc.mp3"
# output_dir = "/content/wavs4"
# split_audio_on_silence(file_path, output_dir)
```

```python
audio_folder = "/content"
import os

#########################################
#            sửa các thông số sau       #
#########################################
version = "wav_p3"
id_code = "b"
#########################################




idx = 0




file_paths = [os.path.join(audio_folder, file) for file in os.listdir(audio_folder) if '.mp3' in file]
output_dirs = [f"/content/{version}/wavs" + str(i) for i in range(len(file_paths))]
with open('train.txt', 'w', encoding='utf-8') as f_train:
  for file_path, output_dir in zip(file_paths, output_dirs):
    print('processing', file_path)
    if output_dir.split('/')[-1] in []:
      continue
    print(file_path, output_dir)
    split_audio_on_silence(file_path, output_dir)
    with open(output_dir+'.txt','w',encoding='utf-8') as f:
      for file in os.listdir(output_dir):
        file_path = os.path.join(output_dir, file)
        text = transcriber(file_path)['text']
        if len(text) > 130:
          continue
        f.write(f'{file}\t{text}\n')
        f_train.write(f'{id_code}{idx:06d}|vietnamese-00|vietnamese|{file_path.split("content/")[1]}|||{text}|\n')
        idx += 1
```
